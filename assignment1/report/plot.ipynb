```python
import math
def KL(x ,y):
    if x == 0 :
        return math.log(1/(1 - y))
    elif x == 1 :
        return math.log(1/y)
    return (x * math.log(x / y) + (1 - x) * math.log((1 - x) / (1 - y)))

LB = []
p2 = np.linspace(0,0.9,19)
p1 = p2 + 0.1
for i in range(len(p2)):
    LB.append(1/KL(p2[i],p1[i]))
plt.plot(p2,LB)
plt.xlabel('$p_2$')
plt.ylabel('$1/KL(p_2,p_1)$')
plt.savefig('../images/task2b_support.png')
plt.show()
```
